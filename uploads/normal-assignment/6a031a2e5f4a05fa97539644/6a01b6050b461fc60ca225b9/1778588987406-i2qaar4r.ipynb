```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Setting a random seed for reproducibility in testing
np.random.seed(42)
```

```python
# Simulating a dataset: Advertising Spend vs Sales
data = {
    'TV_Ads': [230.1, 44.5, 17.2, 151.5, 180.8, 8.7, 57.5, 120.2, 8.6, 199.8],
    'Sales': [22.1, 10.4, 9.3, 18.5, 12.9, 7.2, 11.8, 13.2, 4.8, 10.6]
}
df = pd.DataFrame(data)

X = df['TV_Ads'].values
y = df['Sales'].values

plt.scatter(X, y, color='blue', label='Data Points')
plt.xlabel('TV Advertising Spend')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

## The Core Algorithm (Structural AI Target)

This cell is what your Tree-sitter parser will analyze to create an Abstract Syntax Tree (AST).

```python
def compute_cost(X, y, m, b):
    """Calculates Mean Squared Error."""
    n = len(y)
    total_error = np.sum((y - (m * X + b)) ** 2)
    return total_error / (2 * n)

def gradient_descent(X, y, m, b, learning_rate, epochs):
    """Updates weights m and b to minimize cost."""
    n = len(y)
    cost_history = []

    for i in range(epochs):
        y_pred = m * X + b
        
        # Derivatives
        dm = (-2/n) * np.sum(X * (y - y_pred))
        db = (-2/n) * np.sum(y - y_pred)
        
        # Update parameters
        m = m - (learning_rate * dm)
        b = b - (learning_rate * db)
        
        cost_history.append(compute_cost(X, y, m, b))
        
    return m, b, cost_history
```

```python
# Hyperparameters
L_RATE = 0.00001
EPOCHS = 1000

# Training
m_final, b_final, history = gradient_descent(X, y, 0, 0, L_RATE, EPOCHS)

print(f"Final Weight (m): {m_final}")
print(f"Final Bias (b): {b_final}")

# Plotting the Regression Line
plt.scatter(X, y, color='blue')
plt.plot(X, m_final * X + b_final, color='red', linewidth=3)
plt.title('Linear Regression Result')
plt.show()
```
